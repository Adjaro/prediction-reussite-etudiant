## prediction-reussite-etudiant


```python
import pandas as pd
import  numpy  as  np
import matplotlib.pyplot as plt
```

```python
csv_path = "data/NotesUes.csv"
nrows  = 10000  
df = pd.read_csv(csv_path, nrows=nrows)
```

```python
df.head()
```

```python
df.dtypes
```

```python
resultat_counts = df.groupby('resultat').size()

resultat_counts.plot(kind='bar', title='Répartition des résultats')
plt.xlabel('Résultat')
plt.ylabel("Nombre d'étudiants")

plt.show()
```

```python
df_valide = df[df['resultat'] == 'Validé']
df_valide
```

```python
df_valide['annee'] = df_valide['semestre'].str.extract('(\d{4})')
years = df_valide['annee'].unique()
years = np.sort(years)
years
```

```python
idi = df_valide['id'].unique()
idi = np.sort(idi)
idi
```

```python
df_valide = df_valide.groupby(['id', 'annee']).agg({'credit': 'sum'}).reset_index()

df_valide
```

```python
df_pivot = df_valide.pivot(index='id', columns='annee', values='credit').reset_index()
df_pivot
```

```python
df_pivot = df_pivot.fillna(0)
df_pivot
```

```python
def cal_credits(student):
    total_credits = 0  
    consecutive_zeros = 0  
    nombre_annee = 0
    
    for year in years:
        total_credits += student[year]

        if student[year] != 0:
            nombre_annee+=1 

        
        if student[year] == 0:
            consecutive_zeros += 1
     
        else:
            consecutive_zeros = 0

        if consecutive_zeros >= 3:
            break
    
    return (total_credits,nombre_annee)  


df_pivot['total'] = df_pivot.apply(cal_credits, axis=1)

print(df_pivot)
```

```python
df_pivot.set_index("id", inplace=True)
df_pivot
```

```python
df_pivot['total'].unique()
```

```python
resultat_counts = df_pivot[df_pivot['total'] != (0,0)]['total'].value_counts()
print(resultat_counts )
```

```python
resultat_counts.plot(kind='bar', color= 'orange')
plt.title("Répartition des résultats")
plt.xlabel("Résultat")
plt.ylabel("Nombre d'étudiants")
plt.xticks(rotation=0)
plt.show()
```
